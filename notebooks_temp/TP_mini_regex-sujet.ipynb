## Exemples avec Python

````python
import re
texte = """\
Les sanglots longs des violons de l'automne 
blessent mon cœur d'une langueur monotone. 
Tout suffocant et blême, quand sonne l'heure, 
je me souviens des jours anciens et je pleure
"""

## %
# Capturer tous les o suivis de 3 caractères
for r in re.findall("o...", texte):
    print(r)
````

````python
# Capturer toutes les lignes contenant un b
# re.MULTILINE est nécessaire pour que ^ matche un début de ligne
# et non juste le début de la chaîne.
for r in re.findall("^.*b.*$", texte, re.MULTILINE):
    print(r)
````

````python
# Capturer les chaînes composées d'un o, puis d'une série de lettres (sauf un a ou un passage à la ligne), puis d'un n
for r in re.findall("o[^a\n]*n", texte):
    print(r)
````

````python
# Capturer les mots qui contiennent un `t` majuscule ou minuscule
re.findall("\w*[tT]\w*", texte)
````

On peut créer des groupes dans la chaîne capturée et les extraire

````python
# Capturer les chaînes composées d'un o, puis d'une série d'au moins 3 lettres, 
# puis un a. N'afficher QUE ce qui est entre le o et le a
for r in re.findall("o(....*)a", texte):
    print(r)
````

La fonction `re.findall` permet de renvoyer une liste contenant 
 toutes les chaînes capturées.
 Le motif complet est renvoyé s'il n'y a pas de de groupe. S'il y a
des groupes, ce sont des tuples, formés par ces groupes qui sont renvoyés.

Voici quelques exemples : 

````python
texte2 = "ABCDEFGAEIOU"

print(re.findall(".E.", texte2))
print(re.findall(".(E.)", texte2))
print(re.findall("(.)E(.)", texte2))
````

## Échauffement

La fonction `re.findall` renvoie toutes les chaînes capturées.
Voici quelques questions (la première est corrigée pour que vous voyiez ce qui est attendu) sur les expressions rationnelles.
Les exemples sont exécutés sur le texte suivant :


````python
oiseaux_passage = """\
Oh ! les gens bienheureux !... Tout à coup, dans l'espace,
Si haut qu'il semble aller lentement, un grand vol
En forme de triangle arrive, plane et passe.
Où vont-ils ? Qui sont-ils ? Comme ils sont loin du sol !
Regardez-les passer ! Eux, ce sont les sauvages.
Ils vont où leur désir le veut, par-dessus monts,
Et bois, et mers, et vents, et loin des esclavages.
L'air qu'ils boivent ferait éclater vos poumons.
Regardez-les, vieux coq, jeune oie édifiante !
Rien de vous ne pourra monter aussi haut qu'eux.
Et le peu qui viendra d'eux à vous, c'est leur fiente.
Les bourgeois sont troublés de voir passer les gueux."""
````

### Capturer tous les vers commençant pas une voyelle

````python
re.findall("^[AEIOU].*$", oiseaux_passage, re.M)
````

### Capturer tous les mots commençant par un `m`
Vous pourrez supposer qu'un mot qui commence par une minuscule est nécessairement précédé par un espace.
Vous devez obtenir :

~~~python
[' monts', ' mers', ' monter']
~~~

````python
re.findall(" m\w*",oiseaux_passage)
````

Comme vous l'avez remarqué, la chaîne capturée contient l'espace initial. En utilisant les groupes, faites en sorte de ne récupérer que les mots capturés, sans l'espace initial. Vous devez obtenir :

~~~python
['monts', 'mers', 'monter']
~~~

````python
re.findall(" (m\w*)",oiseaux_passage)
````

### Capturer tous les mots qui sont après `et` ou `Et`
Par chance, il n'y a pas de mot se terminant par `et`, excepté `et` lui même... et le mot `et` n'est jamais suivi d'une ponctuation... Vous ne devez récupérer **que** le mot en question :

~~~python
['passe', 'bois', 'mers', 'vents', 'loin', 'le']
~~~

````python
re.findall("[eE]t (\w*)",oiseaux_passage)
````

### Opérateur non glouton

Ci-dessous, on capture les chaînes commençant par un `e`, et se terminant par un `e`. Comme par défaut l'opérateur `*` est *glouton* (*greedy*), c'est la chaîne la plus longue qui sera capturée... :

````python
re.findall("e.*e", oiseaux_passage)
````

Il y a un moyen d'utiliser une version *non gloutonne* de `*`, de manière à capturer les chaînes les plus courtes : 
`es ge`, `enhe`... Mettez ce moyen en oeuvre. Vous devez obtenir 32 captures (il n'y a que les 5 premières ci-dessous) :

~~~python
['es ge',
 'enhe',
 "eux !... Tout à coup, dans l'e",
 'emble',
 'er le',
 ...
]


````python
re.findall("e.*?e", oiseaux_passage)
````

### Capturer des phrases

En plus de l'option `re.M` que vous avez utilisée, l'option `re.S` permet de traiter le retour à la ligne comme un caractère capturable par `.` (ainsi, on peut capturer avec `.*` des chaînes qui courent sur plusieurs lignes). Par exemples :

````python
re.findall("[A-Z].*?[.?!]",oiseaux_passage,re.S)
````

Utilisez ce que vous avez vu pour capturer les phrases. Une phrase commence
par une majuscule, et se termine par un point (normal, exclamation, interrogation).
Attention, une phrase peut courir sur plusieurs vers. Vous devez capturer 13
phrases :

~~~python
['Oh !',
 "Tout à coup, dans l'espace,\nSi haut qu'il semble aller lentement, un grand vol\nEn forme de triangle arrive, plane et passe.",
 'Où vont-ils ?',
 'Qui sont-ils ?',
 'Comme ils sont loin du sol !',
 'Regardez-les passer !',
 'Eux, ce sont les sauvages.',
 'Ils vont où leur désir le veut, par-dessus monts,\nEt bois, et mers, et vents, et loin des esclavages.',
 "L'air qu'ils boivent ferait éclater vos poumons.",
 'Regardez-les, vieux coq, jeune oie édifiante !',

~~~

Notez aussi que les caractères spéciaux comme `.` ou `*` perdent leur attribut spécial
à l'intérieur de `[ ]`. Par exemple, `[a.*]` signifie bien `a` ou `.` ou `*`.
Sinon, il est possible de rendre un caractère spécial ordinaire en l'échappant avec '\' :
`"\.\w*` matche tous les mots préfixés par un point.

Nous voulons récupérer automatiquement les noms des auteurs pour les afficher, triés par ordre alphabétique (du prénom... :) )

La suite montre comment récupérer le source de la page Web en question et l'afficher. À vous de produire la capture et l'affichage des auteurs (sans les commentaires entre parenthèses qui sont parfois ajoutés après les noms d'auteur).

Récemment, Wikimedia a dû faire par à l'appétit gargantuesque des dresseurs de LLM 😀 qui ont consommé beaucoup de bande passante pour aspirer les pages de l'encyclopédie en ligne.
Afin de freiner ceci, Wikimedia a mis en place un filtre qui essaie de détecter si le traffic vient d'un navigateur ou d'un programme (on dit un robot).
Pour cela, les serveurs Web testent un en-tête HTTP (protocole utilisé pour récupérer les pages Web), et s'il est caractéristique de programmes (curl, python etc.), la page n'est pas servie.

Il est possible de passer outre cette mesure en fournissant un entête caractéristique des navigateurs. Pour cela, il faut remplacer la ligne : 

```python
rpage = requests.get("https://...")
```
par :

```python
rpage = requests.get("https://...", headers={"User-Agent": "Mozilla/5.0 Gecko/20100101 Firefox/147.0")
```

Merci de ne pas en abuser. Il est par exemple inutile de recharger la page 10 fois. Une fois suffit...

</div>

````python
# Récupération et affichage du source de la page)
# C'est ce que vous voyez si vous affichez la page dans un navigateur et faites *Voir le code source de la page*
import requests
rpage = requests.get("https://en.wikipedia.org/wiki/List_of_science_fiction_novels", headers={"User-Agent": "Mozilla/5.0 Gecko/20100101 Firefox/147.0"})
print(rpage.text)
````

Votre travail devrait figurer dans les cellules qui suivent (vous pouvez en ajouter...)

````python
Auteurs = re.findall("by <a href=.*?>(.*?)</a>",rpage.text)
````

**Explication de la réponse ci-dessus :**
Cette réponse utilise une expression régulière (`re.findall`) pour extraire spécifiquement les noms des auteurs dans le code source de la page Wikipédia. L'expression cible les balises `<a>` situées après le mot 'by', et capture grâce aux parenthèses `(.*?)` le texte contenu à l'intérieur du lien HTML (c'est-à-dire le nom de l'auteur).

````python
tri = sorted(set(Auteurs))
````

**Explication de la réponse ci-dessus :**
Cette réponse sert à éliminer les doublons et à trier les noms des auteurs. La fonction `set(Auteurs)` convertit la liste en un ensemble (qui ne peut pas contenir de doublons), puis `sorted()` trie cet ensemble par ordre alphabétique pour obtenir une liste propre et lisible.
