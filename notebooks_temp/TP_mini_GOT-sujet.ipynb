**Combien de lignes contenait le fichier `.csv` ?**

```python
n = len(relations_)
print(n)
```

**Explication de la réponse ci-dessus :**
Cette commande compte et affiche le nombre d'éléments (lignes) contenus dans la liste `relations_`, ce qui donne la taille du jeu de données récupéré depuis le fichier CSV.

Commencez par nettoyer la liste en retirant le premier élément (en-tête), et en faisant en sorte que le troisième élément des triplets soit un nombre et non une chaîne de caractères.

Utiliser les listes en compréhension est assez approprié....

```python
relations = [[p1,p2,int(w)] for p1,p2,w in relations_[1:]]
```

**Explication de la réponse ci-dessus :**
Cette réponse a pour but de nettoyer les données brutes. En utilisant une liste en compréhension avec `relations_[1:]`, on ignore la première ligne (les en-têtes du CSV). De plus, le poids de la relation `w` est converti en entier (`int(w)`) pour pouvoir faire des calculs par la suite.

On désire savoir combien de personnages différents sont mentionnés dans ce fichier.
Pour cela, il faut tout d'abord remarquer qu'un personnage peut apparaître parfois comme premier 
élément et parfois comme second élément d'un triplet. En utilisant les ensembles (`set`) 
construisez la liste des personnages et indiquez combien il y en a.

Dans la suite on appellera $n$ le nombre de lignes du fichier `csv` et $p$ le nombre de personnages différents.
Cette questions doit être réalisée avec une complexité temporelle $\Theta(n)$.

```python
perso1 = {a for a,p,c in relations}
perso2 = {p for a,p,c in relations}
perso = perso1|perso2
print(len(perso))
```

**Explication de la réponse ci-dessus :**
Cette réponse permet de compter le nombre de personnages uniques. On crée deux ensembles (sets) `perso1` et `perso2` regroupant respectivement les personnages de la première et deuxième colonne. L'opérateur `|` réalise l'union des deux ensembles, supprimant ainsi les doublons, et `len()` donne le nombre total.

Assurez-vous maintenant que si 2 personnages `A` et `B` sont mis en relation par une ligne `A, B, x` ils ne figurent pas aussi dans le fichier sous la forme `B, A, y`.
Cette question doit être réalisée avec une complexité temporelle en $\Theta(n)$.

```python
ordre1 = {(a,b) for a,b,c in relations}
ordre2 = {(b,a) for a,b,c in relations}
print(ordre1 & ordre2)
```

**Explication de la réponse ci-dessus :**
Cette réponse vérifie si les relations sont symétriques ou s'il y a des doublons inversés. En créant deux sets (`ordre1` avec les paires `(a,b)` et `ordre2` avec les paires `(b,a)`) et en faisant leur intersection, on vérifie quelles relations existent dans les deux sens dans le fichier.

Pour chaque personnage, déterminez avec combien de personnages différents il a des interactions. Puis, produisez une liste des personnages par importance décroissante selon ce critère. Notez que le troisième paramètre numérique n'est pas utilisé ici. La complexité temporelle doit être $\Theta(n)$ (et non $\Theta(np)$).

**Quels sont les 5 personnages ayant le plus de *connaissances* différentes ?**

```python
scores = {}
for a,b,c in relations:
    scores[a] = scores.get(a,0)+1
    scores[b] = scores.get(b,0)+1

sorted(scores.items(), key=lambda x:x[1], reverse = True)[:5]
```

**Explication de la réponse ci-dessus :**
Cette réponse permet de calculer le nombre total d'interactions pour chaque personnage. En utilisant un dictionnaire `scores`, on incrémente la valeur à chaque fois qu'un personnage apparaît dans une relation. Ensuite, `sorted` est utilisé avec une fonction lambda pour trier le dictionnaire par valeur décroissante et récupérer le top 5.

Compter le nombre de relations est une chose. Mais nous pouvons aussi mesurer l'intensité de ces relations en  comptabilisant le nombre d'interactions.
Ainsi, on peut considérer qu'un personnage qui a 10 interactions avec chacune de ses 2 relations est plus important qu'un personnage qui a une seule interaction avec chacune de ses 3 relations.
